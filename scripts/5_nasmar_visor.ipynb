Tras analizar el posible uso del GeoVisor y los datos proporcionados se plantean varias cuestiones.
La primera y más importante **cual es el objetivo del Geovisor**. Desde iCarto lo definimos como **“Seguimiento de indicadores de desarrollo relacionados con el agua en comunidades rurales donde se levantan datos de Fonsagua en los siete municipios de NASMAR”** Lo que tiene varias implicaciones. Por ejemplo que el poner datos de microcuencas o pozos no encaja con el objetivo.
```python
# Sobre la existencia de Juntas de Agua

# En fonsagua el campo `abastecimientos.h_juntas_agua` indica si para un abastecimiento hay una Junta de Agua.
# También tenemos la tabla `juntas_agua` que tiene un campo que nos indica a que abastecimiento pertenece.
# Y a modo de chequeo podemos generar tanto el campo `h_juntas_agua` como ese conteo:

query = """
SELECT 
    cod_abastecimiento
    , h_juntas_agua
    , COALESCE((SELECT count(*) FROM juntas_agua j WHERE j.cod_abastecimiento = a.cod_abastecimiento GROUP BY j.cod_abastecimiento), 0) as numero_ja
FROM abastecimientos a
ORDER BY cod_abastecimiento;
"""
juntas_agua_check = pandas_utils.get_dataframe_from_sqlite_query(sqlite_path, query, index_col="cod_abastecimiento")
display(juntas_agua_check)

# Como los resultados tiene sentido podemos tirar directamente del valor `h_juntas_agua` que convertiremos a 'Sí', 'No' en lugar de 'true', 'false'/None.

# Para relacionar juntas y comunidades tenemos que pasar por la tabla intermedia de
# `r_abastecimientos_comunidades` que nos dice que comunidades pertenecen a un determinado
# abastecimiento. De modo que diremos que una comunidad tiene junta de aguas, si tiene un
# abastecimiento registrado y ese abastecimiento tiene junta de aguas.

query = """
SELECT
    r.cod_comunidad
    , r.cod_abastecimiento
    , h_juntas_agua
    , COALESCE((SELECT count(*) FROM juntas_agua j WHERE j.cod_abastecimiento = a.cod_abastecimiento GROUP BY j.cod_abastecimiento), 0) as numero_ja
FROM
    r_abastecimientos_comunidades r
    LEFT OUTER JOIN abastecimientos a USING (cod_abastecimiento)
ORDER BY r.cod_comunidad, r.cod_abastecimiento;
"""

juntas_agua_check_con_comunidad = pandas_utils.get_dataframe_from_sqlite_query(sqlite_path, query, index_col="cod_abastecimiento")
display(juntas_agua_check_con_comunidad)

# Lo que nos viene a decir este resultado es que tenemos un total de 37 comunidades
# relacionadas con un total de 28 abastecimientos. En Fonsagua hay registradas 107
# comunidades. Y que esas 37 tienen junta de agua.
```

```python
# Exportamos las capas que usaremos en el visor con sólo los campos que necesitaremos


# Microcuencas
microcuencas["nombre"] = microcuencas["Cuenca"] + " - " + microcuencas["Nombre"]
microcuencas_export = microcuencas.loc[:, ["nombre", "geometry"]]
microcuencas_export.to_crs("EPSG:4326", inplace=True)
microcuencas_export.to_file("microcuencas_visor.shp", encoding="utf-8")
# display(microcuencas_export)


# Municipios NASMAR
municipios_nasmar_export = municipios_nasmar.loc[:, ["nombre", "geom"]]
municipios_nasmar_export.to_crs("EPSG:4326", inplace=True)
municipios_nasmar_export.to_file("municipios_nasmar_visor.shp", encoding="utf-8")
# display(municipios_nasmar_export)


# Comunidades INE
sql = """
    SELECT
        null as c_fonsagua
        , null as n_fonsagua
        , cod_caseri as c_ine
        , caserio as n_ine
        , (SELECT dpto FROM departamentos WHERE substr(c.cod_caseri, 1, 2) == cdpto) as n_depart
        -- , departamen as n_depart
        , (SELECT munic FROM municipios WHERE substr(c.cod_caseri, 1, 4) == cod_munic) as n_muni
        -- , municipio as n_muni
        , (SELECT canton FROM cantones WHERE substr(c.cod_caseri, 1, 6) == cod_canton) as n_aldea
        -- , aldea as n_aldea
        , pob_total as pob_total
        , categoria as categoria
        , null as fam_total
        , null as viv_total
        , null as ano_pgirh
        , null as san_total
        , null as abas_total
        , null as junta_agua
        , ST_AsBinary(geom) AS geom
    FROM
        caserios_comunidades c
"""
comunidades_ine_export = get_dataframe_from_spatialite_query(sqlite_path, sql, index_col=None, crs={"init": "epsg:32616"})
nasmar.adjust_code_to_9_from_8(comunidades_ine_export, "c_ine")
nasmar.drop_elements_not_in_nasmar_councils(comunidades_ine_export, "c_ine", raise_error=False, log_warn=False)
comunidades_ine_export.sort_values(by="c_ine", axis="index", ascending=True, inplace=True)
comunidades_ine_export.to_crs("EPSG:4326", inplace=True)
comunidades_ine_export.to_file("comunidades_ine_visor.shp", encoding="utf-8")
# display(comunidades_ine_export)
# comunidades_ine_export.plot()


# Comunidades FONSAGUA
sql = """

    WITH tmp_juntas_agua AS (
        SELECT
            r.cod_comunidad
            , r.cod_abastecimiento
            , h_juntas_agua
            , CASE WHEN h_juntas_agua == 'true' THEN 'Sí' ELSE 'No' END as result
            , COALESCE((SELECT count(*) FROM juntas_agua j WHERE j.cod_abastecimiento = a.cod_abastecimiento GROUP BY j.cod_abastecimiento), 0) as numero_ja
         FROM
             r_abastecimientos_comunidades r
             LEFT OUTER JOIN abastecimientos a USING (cod_abastecimiento)
          ORDER BY r.cod_comunidad, r.cod_abastecimiento
    )

    SELECT
        cod_comunidad as c_fonsagua
        , comunidad as n_fonsagua
        , cod_ine_2013 as c_ine
        , null as n_ine
        , (SELECT dpto FROM departamentos WHERE substr(c.cod_comunidad, 1, 2) == cdpto) as n_depart
        -- , departamento as n_depart
        , (SELECT munic FROM municipios WHERE substr(c.cod_comunidad, 1, 4) == cod_munic) as n_muni
        -- , municipio as n_muni
        , (SELECT canton FROM cantones WHERE substr(c.cod_comunidad, 1, 6) == cod_canton) as n_aldea
        -- , canton as n_aldea
        , n_habitantes as pob_total
        , null as categoria
        , n_familias as fam_total
        , n_viviendas as viv_total
        , ano_pgirh
        , min(let_vivienda, n_viviendas) as san_total
        , min(n_familias, max(tot_s_abast, tot_ll_abast)) as abas_total
        , COALESCE((SELECT result FROM tmp_juntas_agua j WHERE j.cod_comunidad = c.cod_comunidad), 'No') as junta_agua
        , ST_AsBinary(geom) AS geom
    FROM 
        comunidades c
    WHERE cod_comunidad NOT IN ('06072302', '06060399')
"""

comunidades_fonsagua_export = get_dataframe_from_spatialite_query(sqlite_path, sql, index_col=None, crs={"init": "epsg:32616"})
nasmar.adjust_code_to_9_from_8(comunidades_fonsagua_export, "c_fonsagua")
nasmar.adjust_code_to_9_from_8(comunidades_fonsagua_export, "c_ine")
nasmar.drop_elements_not_in_nasmar_councils(comunidades_fonsagua_export, "c_fonsagua", raise_error=False, log_warn=False)
comunidades_fonsagua_export.sort_values(by="c_fonsagua", axis="index", ascending=True, inplace=True)
comunidades_fonsagua_export.to_crs("EPSG:4326", inplace=True)
comunidades_fonsagua_export.to_file("comunidades_fonsagua_visor.shp", encoding="utf-8")
# display(comunidades_fonsagua_export)


# Comunidades del INE que no tienen datos en Fonsagua más Comunidades Fonsagua
sql = """
    WITH comunidades_ine AS (
        SELECT
            -- null as c_fonsagua
            -- , null as n_fonsagua
            cod_caseri as codigo
            , caserio as nombre
            , (SELECT dpto FROM departamentos WHERE substr(cc.cod_caseri, 1, 2) == cdpto) as n_depart
            , (SELECT munic FROM municipios WHERE substr(cc.cod_caseri, 1, 4) == cod_munic) as n_muni
            , (SELECT canton FROM cantones WHERE substr(cc.cod_caseri, 1, 6) == cod_canton) as n_aldea
            , pob_total as pob_total
            , categoria as categoria
            , null as fam_total
            , null as viv_total
            , null as ano_pgirh
            , null as san_total
            , null as abas_total
            , null as junta_agua
            , ST_AsBinary(geom) AS geom
        FROM
             caserios_comunidades cc
        WHERE
             cod_caseri NOT IN (
                 SELECT cod_comunidad FROM comunidades WHERE cod_comunidad NOT IN ('06072302', '06060399')
                 UNION
                 SELECT cod_ine_2013 FROM comunidades WHERE cod_ine_2013 IS NOT NULL AND cod_comunidad NOT IN ('06072302', '06060399')
             )
    )
    , tmp_juntas_agua AS (
        SELECT
            r.cod_comunidad
            , r.cod_abastecimiento
            , h_juntas_agua
            , CASE WHEN h_juntas_agua == 'true' THEN 'Sí' ELSE 'No' END as result
            , COALESCE((SELECT count(*) FROM juntas_agua j WHERE j.cod_abastecimiento = a.cod_abastecimiento GROUP BY j.cod_abastecimiento), 0) as numero_ja
         FROM
             r_abastecimientos_comunidades r
             LEFT OUTER JOIN abastecimientos a USING (cod_abastecimiento)
          ORDER BY r.cod_comunidad, r.cod_abastecimiento
    )
    SELECT
        cod_comunidad as codigo
        , comunidad as nombre
        -- , cod_ine_2013 as c_ine
        -- , null as n_ine
        , (SELECT dpto FROM departamentos WHERE substr(c.cod_comunidad, 1, 2) == cdpto) as n_depart
        , (SELECT munic FROM municipios WHERE substr(c.cod_comunidad, 1, 4) == cod_munic) as n_muni
        , (SELECT canton FROM cantones WHERE substr(c.cod_comunidad, 1, 6) == cod_canton) as n_aldea
        , n_habitantes as pob_total
        , (SELECT categoria FROM caserios_comunidades cas WHERE c.cod_comunidad == cas.cod_caseri) as categoria
        , n_familias as fam_total
        , n_viviendas as viv_total
        , ano_pgirh
        , min(let_vivienda, n_viviendas) as san_total
        , min(n_familias, max(tot_s_abast, tot_ll_abast)) as abas_total
        , COALESCE((SELECT result FROM tmp_juntas_agua j WHERE j.cod_comunidad = c.cod_comunidad), 'No') as junta_agua
        , ST_AsBinary(geom) AS geom
    FROM 
        comunidades c
    WHERE cod_comunidad NOT IN ('06072302', '06060399')
    
    UNION 
    SELECT * FROM comunidades_ine
"""
comunidades_export = get_dataframe_from_spatialite_query(sqlite_path, sql, index_col=None, crs={"init": "epsg:32616"})
nasmar.adjust_code_to_9_from_8(comunidades_export, "codigo")
nasmar.drop_elements_not_in_nasmar_councils(comunidades_export, "codigo", raise_error=False, log_warn=False)
comunidades_export.sort_values(by="codigo", axis="index", ascending=True, inplace=True)
comunidades_export.to_crs("EPSG:4326", inplace=True)
comunidades_export.to_file("comunidades_visor.shp", encoding="utf-8")
# display(comunidades_fonsagua_export)
```

# Listado de shapes proporcionados

Todos los shapes usan como encoding UTF-8 y han sido reproyectados a EPSG:4326

* microcuencas_visor.shp
* municipios_nasmar_visor.shp. Los 7 municipios de NASMAR
* comunidades_ine_visor.shp. Las comunidades del INE, con los campos relevantes para el visor a NULL cuando no hay datos para ellos
* comunidades_fonsagua_visor.shp. Las comunidades de Fonsagua
* comunidades_visor.shp. Es la unión de las comunidades del INE que no están ya recogidas en Fonsagua, más las comunidades de Fonsagua

Hay que tener en cuenta que los datos tanto del INE como de Fonsagua se han tomado en distintas fechas, así que en ocasiones comparar la población de una comunidad con otro no tiene sentido por ser de años muy distintos.

Explicación de los campos:

* codigo: Código de la comunidad. Se respetan los códigos de Fonsagua, sobre los del INE, pero sólo son distintos en unos 5 casos.
* nombre: Nombre de la comunidad. Se respetan los nombres usados en Fonsagua sobre los del INE
* n_depart: Nombre del Departamento. Unificado a partir de la capa "departamentos"
* n_muni: Nombre del Municipio. Unificado a partir de la capa "municipios"
* n_aldea: Nombre de la Aldea. Unificado a partir de la capa "cantones"
* pob_total: Población total. Se corresponde al campo "n_habitantes" de Fonsagua
* categoria: Campo proveniente de la capa de comunidades del INE (se ha rellenado en las comunidades de Fonsagua)
* fam_total: Número de familias, campo "n_familias" (que está siempre relleno) de Fonsagua. A nulo para las comunidades del INE
* viv_total: Número de familias, campo "n_viviendas" (que está siempre relleno) de Fonsagua. A nulo para las comunidades del INE
* ano_pgirh: Año de realización del PGIRH. Siempre relleno (excepto 2 comunidades) de Fonsagua. A nulo para las comunidades del INE
* san_total. Número de letrinas en Fonsagua, campo "let_vivienda", corregido para que nunca sea superior al número de viviendas. A nulo para las comunidades del INE
* abas_total. Número de familias con abastecimiento de agua, campos "tot_s_abast" o "tot_ll_abast" de Fonsagua. Corregido para que nunca sea superior al número de viviendas. A nulo para las comunidades del INE
* junta_agua. Si la comunidad tienen o no Junta de Agua.
